# **What is bias?**<br>
# Theoretical explanation
To put in a simple way, biases are just extra inputs we provide for our neurons; therefore, every neuron has its own bias value. Biases are additional parameters in our neural network which are used to adjust the output along with the weighted sum of the inputs.<br>
Conceptually, bias has a role similar to that of a threshold which we will discuss later.<br>
Biases are learnable, just like the weights; so, we can update the values of our biases via the same methods we use to update the weight values.
# Mathematical explanation
Now, let us discuss the role of bias, mathematically.<br>
Technically, bias is a parameter that can shift our activation function. when we are working with linear regression, bias acts just like an intercept added to out linear equation; this means that our graph does not necessarily pass through the origin $(0,0)$.<br><br>
$y = mx + b$

```python
import numpy as np
```

```python
def unit_step_func(x):
  return np.where(x> 0, 1, 0)
```

Let us first not include the bias in our calculations

We are going to create a class, Perceptron and define the fit and predict methods for it.

```python
class Perceptron:
    #lr: learning rate, between 0 and 1
    #n_iters: number of iterations
    def __init__(self, lr = 0.01, n_iters=10):
        self.lr = lr
        self.n_iters = n_iters
        self.activation_func = unit_step_func
        self.weights = None
        # self.bias = 0

    def fit(self, X, y):
        #assuming that X is a numpy nd array
        n_samples, n_features = X.shape
        #init parameters, a better way is to randomly initialize the weights
        self.weights = np.zeros(n_features)
        # self.bias = 0

        y_ = np.where(y>0, 1, 0)

        for _ in range(self.n_iters):
          for indx, x_i in enumerate(X):
            # linear_output = np.dot(x_i, self.weights) + self.bias
            linear_output = np.dot(x_i, self.weights)

            y_pred = self.activation_func(linear_output)

            #Perceptron update rule

            update = self.lr * (y_[indx] - y_pred)

            self.weights += update * x_i
            #self.bias += update

    def predict(self, X):
      #linear_output = np.dot(x_i, self.weights) + self.bias
      linear_output = np.dot(X, self.weights)

      y_pred = self.activation_func(linear_output)
      return y_pred
```

Now, let us test this code.

```python
from sklearn.model_selection import train_test_split
from sklearn import datasets
```

```python
def accuracy(y_true, y_pred):
  accuracy = np.sum(y_true == y_pred) / len(y_true)
  return accuracy*100
```

```python
X, y = datasets.make_blobs(n_samples = 150, n_features = 2, centers = 2, cluster_std = 1.05, random_state = 2)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123)
```

```python
p = Perceptron(lr = 0.01, n_iters = 1)
p.fit(X_train, y_train)
predictions = p.predict(X_test)

print("Perceptron classification accuracy: ", accuracy(y_test, predictions))
```

Now, we will plot the results.

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.scatter(X_train[:, 0], X_train[:, 1], marker="o", c=y_train)

x0_1 = np.amin(X_train[:, 0])
x0_2 = np.amax(X_train[:, 0])

x1_1 = (-p.weights[0] * x0_1) / p.weights[1]
x1_2 = (-p.weights[0] * x0_2) / p.weights[1]

ax.plot([x0_1, x0_2], [x1_1, x1_2], "k")

ymin = np.amin(X_train[:, 1])
ymax = np.amax(X_train[:, 1])
ax.set_ylim(ymin - 3, ymax + 3)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.scatter(X_test[:, 0], X_test[:, 1], marker="o", c=y_test)

x0_1 = np.amin(X_test[:, 0])
x0_2 = np.amax(X_test[:, 0])

x1_1 = (-p.weights[0] * x0_1) / p.weights[1]
x1_2 = (-p.weights[0] * x0_2) / p.weights[1]

ax.plot([x0_1, x0_2], [x1_1, x1_2], "k")

ymin = np.amin(X_test[:, 1])
ymax = np.amax(X_test[:, 1])
ax.set_ylim(ymin - 3, ymax + 3)

plt.show()
```

```python
class Perceptron_with_bias:
    #lr: learning rate, between 0 and 1
    #n_iters: number of iterations
    def __init__(self, lr = 0.01, n_iters=1):
        self.lr = lr
        self.n_iters = n_iters
        self.activation_func = unit_step_func
        self.weights = None
        self.bias = 0

    def fit(self, X, y):
        #assuming that X is a numpy nd array
        n_samples, n_features = X.shape
        #init parameters, a better way is to randomly initialize the weights
        self.weights = np.zeros(n_features)
        self.bias = 0

        y_ = np.where(y>0, 1, 0)

        for _ in range(self.n_iters):
          for indx, x_i in enumerate(X):
            linear_output = np.dot(x_i, self.weights) + self.bias
            # linear_output = np.dot(x_i, self.weights)

            y_pred = self.activation_func(linear_output)

            #Perceptron update rule

            update = self.lr * (y_[indx] - y_pred)

            self.weights += update * x_i
            self.bias += update

    def predict(self, X):
      linear_output = np.dot(X, self.weights) + self.bias
      # linear_output = np.dot(X, self.weights)

      y_pred = self.activation_func(linear_output)
      return y_pred
```

```python
p2 = Perceptron_with_bias(lr = 0.01, n_iters = 1)
p2.fit(X_train, y_train)
predictions = p2.predict(X_test)

print("Perceptron_with_bias classification accuracy: ", accuracy(y_test, predictions),"%")
```
